```python
import pandas as pd
import joblib

df=pd.read_csv('titanic_train.csv')
```

```python
df.isnull().sum()
```

```python
#Llenar los nulos de Cabin con una nueva categoría 'Desconocida'
df['Cabin'] = df['Cabin'].fillna('Desconocida')
```

```python
# Llenar los nulos de Embarked con la moda
moda_embarked = df['Embarked'].mode()[0]
print(f"Moda utilizada para llenar 'Embarked': {moda_embarked}")
df['Embarked'] = df['Embarked'].fillna(moda_embarked)
```

```python
df.info()
```

```python
from sklearn.preprocessing import OrdinalEncoder

# Columnas categóricas a convertir
columnas_categoricas = ['Name','Sex', 'Embarked', 'Cabin', 'Ticket']

encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)

df[columnas_categoricas] = encoder.fit_transform(df[columnas_categoricas])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df_numericas = df.select_dtypes(include=['int64', 'float64'])

matriz_correlacion = df_numericas.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(matriz_correlacion, annot=True, cmap='coolwarm', fmt=".2f", vmin=-1, vmax=1 )
plt.title("Matriz de Correlación")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

atributos = ['Pclass', 'Fare', 'SibSp', 'Parch', 'Cabin']

df_train = df[df['Age'].notnull()]
df_pred = df[df['Age'].isnull()]

# Variables X e y
X_train = df_train[atributos]
y_train = df_train['Age']
X_pred = df_pred[atributos]

# Entrenamiento del modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(X_train, y_train)

# Predecir los valores faltantes
predicciones = modelo.predict(X_pred)

# Rellenar los valores nulos en Age
df.loc[df['Age'].isnull(), 'Age'] = predicciones
```

```python
df.info()
```

```python
X = df.drop(columns=['Survived','PassengerId','Name'])
y = df['Survived']
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import  LogisticRegression
from sklearn.feature_selection import RFE
# Selección de características con RFE
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
modelo_lr = LogisticRegression(solver='lbfgs', max_iter=1000)
rfe = RFE(estimator=modelo_lr, n_features_to_select=4)
rfe.fit(X_scaled, y)
caracteristicas_seleccionadas = X.columns[rfe.support_]
print("Columnas seleccionadas:", caracteristicas_seleccionadas.tolist())
X_seleccionado = X[caracteristicas_seleccionadas]
```

```python
# Escalar los datos seleccionados
scaler = StandardScaler()
X_std = scaler.fit_transform(X_seleccionado)
X_std_df = pd.DataFrame(X_std, columns=X_seleccionado.columns)
```

```python
from sklearn.model_selection import train_test_split
# Dividir en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_std, y, test_size=0.2, random_state=42)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=4)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
```

```python
principalDf = pd.DataFrame(data=X_train_pca[:, :4], columns=[f'PC{i+1}' for i in range(4)])
principalDf = pd.concat([principalDf, y_train.reset_index(drop=True)], axis=1)
principalDf['Survived'] = principalDf['Survived'].astype(str)

print("\nInformación del DataFrame de componentes principales:")
print(principalDf.info())
print(principalDf.head())
```

```python
loadings = pca.components_
loadings_df = pd.DataFrame(loadings, columns=X_seleccionado.columns, index=[f'PC{i+1}' for i in range(4)])

print("\nLoadings de los componentes principales:")
print(loadings_df)
```

```python
import numpy as np
# Varianza explicada por cada componente
explained_variances = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variances)
print("\nVarianza explicada por cada componente:")
print(explained_variances)
print("\nVarianza acumulada:")
print(cumulative_variance)
```

```python
# Verificar cuántas componentes capturan el 95% de la varianza
pca_full = PCA()
pca_full.fit(X_train)
cumulative_variance_full = np.cumsum(pca_full.explained_variance_ratio_)
num_components_for_95_variance = np.argmax(cumulative_variance_full >= 0.95) + 1
print(f"\nSe necesitan {num_components_for_95_variance} componentes para capturar el 95% de la variabilidad.")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
import numpy as np

# --- Random Forest ---
clf_rf = RandomForestClassifier(random_state=42)
clf_rf.fit(X_train_pca, y_train)
score_rf = clf_rf.score(X_test_pca, y_test)
scores_rf = cross_val_score(clf_rf, X_train_pca, y_train, cv=5, scoring='accuracy')

print("\n🌲 [Random Forest]")
print("Accuracy con PCA (5 componentes):", score_rf)
print("Accuracies con validación cruzada:", scores_rf)
print("Promedio:", np.mean(scores_rf))
print("Varianza:", np.var(scores_rf))
print("Desviación estándar:", np.std(scores_rf))

# --- Support Vector Machine (SVM) ---
clf_svm = SVC(kernel='rbf', C=1.0, gamma='scale', random_state=42)
clf_svm.fit(X_train_pca, y_train)
score_svm = clf_svm.score(X_test_pca, y_test)
scores_svm = cross_val_score(clf_svm, X_train_pca, y_train, cv=5, scoring='accuracy')

print("\n🤖 [SVM - Support Vector Machine]")
print("Accuracy con PCA (5 componentes):", score_svm)
print("Accuracies con validación cruzada:", scores_svm)
print("Promedio:", np.mean(scores_svm))
print("Varianza:", np.var(scores_svm))
print("Desviación estándar:", np.std(scores_svm))

# --- Árbol de Decisión ---
clf_tree = DecisionTreeClassifier(random_state=42)
clf_tree.fit(X_train_pca, y_train)
score_tree = clf_tree.score(X_test_pca, y_test)
scores_tree = cross_val_score(clf_tree, X_train_pca, y_train, cv=5, scoring='accuracy')

print("\n🌳 [Árbol de Decisión]")
print("Accuracy con PCA (5 componentes):", score_tree)
print("Accuracies con validación cruzada:", scores_tree)
print("Promedio:", np.mean(scores_tree))
print("Varianza:", np.var(scores_tree))
print("Desviación estándar:", np.std(scores_tree))

# --- K-Nearest Neighbors ---
clf_knn = KNeighborsClassifier(n_neighbors=5)
clf_knn.fit(X_train_pca, y_train)
score_knn = clf_knn.score(X_test_pca, y_test)
scores_knn = cross_val_score(clf_knn, X_train_pca, y_train, cv=5, scoring='accuracy')

print("\n👣 [K-Nearest Neighbors (KNN)]")
print("Accuracy con PCA (5 componentes):", score_knn)
print("Accuracies con validación cruzada:", scores_knn)
print("Promedio:", np.mean(scores_knn))
print("Varianza:", np.var(scores_knn))
print("Desviación estándar:", np.std(scores_knn))
```

```python
joblib.dump(encoder, 'encoder.pkl')
joblib.dump(modelo, 'modelo_regresion_edad.pkl')
joblib.dump(scaler, 'scaler.pkl')
joblib.dump(rfe, 'rfe.pkl')
joblib.dump(pca, 'pca.pkl')

# Guardar los clasificadores
joblib.dump(clf_rf, 'modelo_random_forest.pkl')
joblib.dump(clf_svm, 'modelo_svm.pkl')
joblib.dump(clf_tree, 'modelo_arbol_decision.pkl')
joblib.dump(clf_knn, 'modelo_knn.pkl')
```
